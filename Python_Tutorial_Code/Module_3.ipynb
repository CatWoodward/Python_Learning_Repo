```python
# this is to convert the Period field to date time. Note Period is case sensitive
#https://strftime.org/ for list of date types

df['Period'] = pd.to_datetime(df['Period'], format="%Y-%m") # the dash is the thing that seperates them in the csv e.g. 1920-01

df
```

```python

df['Period'] = pd.to_datetime(df['Period'], format="%Y-%m")

child_identifiers['PersonBirthDate'] = pd.to_datetime(child_identifiers['PersonBirthDate'], format="%Y-%m-%d")
child_identifiers['ExpectedPersonBirthDate'] = pd.to_datetime(child_identifiers['ExpectedPersonBirthDate'], format="%Y-%m-%d")

child_identifiers.info()
```

```python


house_prices_sum = df['Average price All property types'].sum() #.mean or .median would give averages

#print(house_prices_sum)

print(f'The sum of all house prices is {int(house_prices_sum)}.') # could add int after the first curly bracket so it rounds up, as in integar 

print(f'the lowest house price since 1980 was {int(df["Average price All property types"].min())}.')

```

```python
#df["Age of data"] = pd.to_datetime('today') #gives todays date
df["Age of data"] = pd.to_datetime('today') - df['Period']

#df['Age of data'].info() # to see the data type it's assigned.  in this example it's done date time but we want age 

df['Age of data'] = df['Age of data'] / pd.Timedelta(365.25, 'd')
df['Age of data'] = df['Age of data'].astype('int') #int rounds to nearest number

df['Age of data']
```

```python
child_identifiers['Age'] = pd.to_datetime('today') - child_identifiers['PersonBirthDate']   #add a collumn named age to the child identifiers dataset then take today and minus the dob
child_identifiers['Age'] = child_identifiers['Age'] / pd.Timedelta(365.25, "d") # age divided by 365 in days
child_identifiers["Age"] = child_identifiers ["Age"].astype('int') # change the output to an integar whole number

child_identifiers

# instead of today you could add a date ("01/04/2024", format "%d/%m/%y")
```
